# Visualize is there is any missing values in the dataset based on heatmap

```python
import seaborn as sns
sns.heatmap(df.isnull(), annot=False, cmap='viridis')
```

# find out the correlation between variables in the given dataset correlation between different values

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),cbar=True,annot=True,cmap='Reds')
```

# Analyse how the income of a patient affects the no. of visits to the hospital

```python
plt.figure(figsize=(10,10))
plt.scatter(x='income',y='visits',data=df)
plt.xlabel('income')
plt.ylabel('visits')
```

# Count and visualize the number of males and females affected by illness

```python
sns.histplot(df.gender,bins=2)
```

# Visualize the percentage of getting govt health insurance due to low income, due to old age and also the percentage of people having private health isurance

```python
label=['yes','no']
Y=df[df['freepoor']=='yes']
N=df[df['freepoor']=='no']
x=[Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title("Percentage of people getting govt health insurance due to low income")
plt.show()
 
label=['yes','no']
Y=df[df['private']=='yes']
N=df[df['private']=='no']
x=[Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title("Percentage of people having private healhealth insurance")
plt.show()

Y=df[df['freerepat']=='yes']
N=df[df['freerepat']=='no']
x=[Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title("Percentage of people getting govt health insurance due to old age,disability or veteran status")
plt.show()
```

# plot a horizontal bar chart to analyze the reduced days of activity due to illness based on gender

```python
db = df.groupby('gender')['illness'].sum().to_frame().reset_index()
plt.barh(db['gender'],db['illness'],color=['cornflowerblue','lightseagreen'])
plt.title('Bar Chart')
plt.xlabel('Gender')
plt.ylabel('illness')
plt.show()
```
